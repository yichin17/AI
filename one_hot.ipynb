```python
import numpy as np
import pandas as pd

# 定義資料
country = ['Taiwan', 'Australia', 'Ireland', 'Australia', 'Ireland', 'Taiwan']
age = [25, 30, 45, 35, 22, 36]
salary = [20000, 32000, 59000, 60000, 43000, 52000]

# 建立字典
dic = {'Country': country, 'Age': age, 'Salary': salary}

# 使用字典建立 DataFrame
data = pd.DataFrame(dic)

# 顯示 DataFrame
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder() # 使用 LabelEncoder 進行編碼
print(data, "\n")
data_le=pd.DataFrame(dic)
print(data_le, "\n")
data_le['Country'] = labelencoder.fit_transform(data['Country']) #['Country']轉成數字
data_le
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# 創建 ColumnTransformer 實例，用於對資料進行轉換
# 將第 0 列 (Country 列) 使用 OneHotEncoder 進行獨熱編碼，
# remainder='passthrough' 表示保留其餘的列不進行轉換
ct = ColumnTransformer(
    [("Country", OneHotEncoder(), [0])],  # 對第 0 列 (Country 列) 使用 OneHotEncoder
    remainder='passthrough'  # 保留其餘的列
)
print(data_le)
# 使用 ColumnTransformer 對資料框 data_le 進行編碼
data_str_ohe = ct.fit_transform(data_le)

# 將編碼後的結果轉換為 DataFrame 並顯示
pd.DataFrame(data_str_ohe) #0='Australia',1='Ireland',2='Taiwan'
```

```python
# 使用 pd.get_dummies() 函數對資料框 data 進行獨熱編碼
# 這將對資料框中的所有類別變數（categorical variables）進行獨熱編碼，
# 將其轉換為二進制的特徵（0 或 1）
print(data)
data_dum = pd.get_dummies(data)

# 將編碼後的結果轉換為 DataFrame 並顯示
pd.DataFrame(data_dum)
```
